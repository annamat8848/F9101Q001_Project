```python
# Unione dei tre dataset

shanghai_recent_slim.rename(columns = {'world_rank':'world_rank_shanghai','year':'year_shanghai'},inplace=True)

merged=pd.merge(pd.merge(cwur_recent_slim,times_recent_slim,on='UNI_ID',suffixes=('_cwur', '_times'),how='outer'),
                shanghai_recent_slim,
                on='UNI_ID',how='outer')

merged.head()
```

```python
print(merged.dtypes)
```

```python
# Funzione che restituisce la posizione media di un'universit quando il suo ranking  espresso come intervallo
# Da applicare a world_rank_shanghai e world_rank_times 

def get_mean_range(element):
    
    if isinstance(element, str) :
    
        element = element.strip('=')
        element = element.split('-')

        if len(element) == 2:
            return (int(element[0])+int(element[1]))/2
        else:
            return int(element[0])     
    
    else:
        return element
```

```python
prova = '201-250'
get_mean_range(prova)
```

```python
merged['world_rank_shanghai_num']=merged['world_rank_shanghai'].apply(get_mean_range)
```

```python
merged['world_rank_times_num']=merged['world_rank_times'].apply(get_mean_range)
```

```python
merged.head()
```

```python
# Restituisce la differenza maggiore tra i ranking di un'universit

def max_rank_dif(row):
    return( max(row['world_rank_cwur'],row['world_rank_shanghai_num'],row['world_rank_times_num']) 
        - min(row['world_rank_cwur'],row['world_rank_shanghai_num'],row['world_rank_times_num']))
```

```python
merged['max_rank_dif']=merged.apply(max_rank_dif,axis=1)

# Delete Nan
merged = merged.dropna(subset=["max_rank_dif"])
merged.head()
```

### 5.Consider only the most recent data point of the times dataset. Compute the number of male and female students for each country.

```python
print(times_recent.dtypes)
```

```python
# Need a numeric variable
# Alcune migliaia specificate con la virgola

times_recent['num_students'].replace(',','', inplace=True,regex=True)
times_recent['num_students']=pd.to_numeric(times_recent['num_students'])
print(times_recent.dtypes)
```

```python
times_recent['female_male_ratio'].replace(r'-',np.nan, inplace=True)
times_recent.head()
```

```python
# Espressione regolare che inidentifica due gruppi: percentuale femmine e maschi

ratio = re.compile('^(?P<fem>\d+) : (?P<male>\d+)$')

prova = ratio.search('48 : 52')
prova.group('fem')
```

```python
# Funzione che applica l'espressione regolare "ratio"

def get_genders_percent(element, gender):
    
    if isinstance(element, str) :
    
        found = ratio.search(element)

        if found:
            if gender == 'F':
                return int(found.group('fem'))/100

            if gender == 'M':
                return int(found.group('male'))/100
    
    else:
        return np.nan
```

```python
# Proporzione
times_recent['females_p'] = times_recent['female_male_ratio'].apply(get_genders_percent, args=('F'))
times_recent['males_p'] = times_recent['female_male_ratio'].apply(get_genders_percent, args=('M'))
times_recent.head()
```

```python
times_recent['females_num'] = round(times_recent['females_p']*times_recent['num_students'])
times_recent['males_num'] = round(times_recent['males_p']*times_recent['num_students'])
times_recent.head()
```

#### 13. Soluzione alternativa

```python
merged_full['world_rank_shanghai']=merged_full['world_rank_shanghai'].str.strip('=')
merged_full['world_rank_shanghai_list']=merged_full['world_rank_shanghai'].str.split('-')

merged_full['world_rank_times']=merged_full['world_rank_times'].str.strip('=')
merged_full['world_rank_times_list']=merged_full['world_rank_times'].str.split('-')

merged_full.head()
```

```python
def rank_compare(row):
    
    cwur=row['world_rank_cwur']
    shanghai=row['world_rank_shanghai_list']
    times=row['world_rank_times_list']
    
    
    if len(shanghai)==2 and len(times)==2:
        
        if(
        int(shanghai[0]) <= cwur <= int(shanghai[1]) and 
        int(times[0]) <= cwur <= int(times[1])
        ):
            
            return True
        else:
        
            return False 
    
    
    if len(shanghai)==1 and len(times)==2:
        
        if( 
        int(shanghai[0])==cwur and 
        int(times[0]) <= cwur <= int(times[1])
        ):
                
            return True
        else:
        
            return False 
    
            
    if len(times)==1 and len(shanghai)==2:
        
        if(        
        int(times[0])==cwur and             
        int(shanghai[0]) <= cwur <= int(shanghai[1])
        ):
                
            return True
        else:
        
            return False 
    
    
    if len(times)==1 and len(shanghai)==1:
    
        if int(times[0])==cwur and int(shanghai[0])==cwur:
        
            return True
        else:
        
            return False 
```

```python
merged_full['rank_compare']=merged_full.apply(rank_compare ,axis=1)
```
